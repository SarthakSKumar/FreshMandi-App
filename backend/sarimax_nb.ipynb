```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
import numpy as np

# Load your full dataset
# Assuming `df` is your full DataFrame with 'date', 'price', and 'precipitation' columns

# Ensure 'date' is in datetime format and set as index
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Split the data chronologically 80-20 for train-test
train_size = int(len(df) * 0.8)
train_df = df.iloc[:train_size]
test_df = df.iloc[train_size:]

# Separate the target variable ('price') and the exogenous variable ('precipitation')
train_y = train_df['price']
train_exog = train_df[['precipitation']]

test_y = test_df['price']
test_exog = test_df[['precipitation']]

# Define SARIMAX model parameters
model = SARIMAX(train_y, exog=train_exog, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))

# Fit the SARIMAX model
sarimax_model = model.fit(disp=False)

# Forecast using the test exogenous variable data
forecast = sarimax_model.get_forecast(steps=len(test_y), exog=test_exog)
predictions = forecast.predicted_mean

# # Combine the date, actual price, and predicted price in a DataFrame for easy comparison
# results_df = pd.DataFrame({
#     'date': test_df.index,
#     'actual_price': test_y,
#     'predicted_price': predictions
# }).set_index('date')

# print("Predictions with Dates:")
# print(results_df)

# Manually print the date, actual price, and predicted price
print("Predictions with Dates:")
for date, actual, predicted in zip(test_df.index, test_y, predictions):
    print(f"Date: {date.strftime('%Y-%m-%d')}, Actual Price: {actual}, Predicted Price: {predicted}")



# Calculate RMSE, MAPE
rmse = np.sqrt(mean_squared_error(test_y, predictions))
mape = mean_absolute_percentage_error(test_y, predictions)

print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Percentage Error (MAPE): {mape * 100:.2f}%")
```
